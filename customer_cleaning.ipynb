**Demographic Information:**

*   **Gender:** Gender of the passenger (Male/Female).
*   **Age:** Age of the passenger.
*   **Customer Type:** Type of customer (Loyal customer/disloyal customer).
*   **Type of Travel:** Purpose of travel (Personal Travel/Business travel).
*   **Class:** Travel class (Business/Eco/Eco Plus).

**Service and Experience Ratings (on a scale of 0-5, where higher is better):**

*   **Inflight wifi service:** Quality of inflight wifi service.
*   **Departure/Arrival time convenient:** Convenience of departure/arrival times.
*   **Ease of Online booking:** Ease of online booking.
*   **Food and drink:** Quality of food and drinks.
*   **Online boarding:** Ease of online boarding.
*   **Seat comfort:** Comfort of the seat.
*   **Inflight entertainment:** Quality of inflight entertainment.
*   **On-board service:** Quality of on-board service.
*   **Leg room service:** Leg room available.
*   **Baggage handling:** Quality of baggage handling.
*   **Checkin service:** Efficiency of check-in service.
*   **Inflight service:** Quality of inflight service.
*   **Cleanliness:** Cleanliness of the aircraft.

**Flight Information:**

*   **Flight Distance:** Distance of the flight.
*   **Departure Delay in Minutes:** Departure delay in minutes.
*   **Arrival Delay in Minutes:** Arrival delay in minutes.

**Target Variable:**

*   **satisfaction:** Overall satisfaction of the passenger (satisfied/neutral or dissatisfied).

**Age Distribution:** Most passengers are aged between 27 and 51, with an average age of 39.  This suggests a focus on a potentially working/middle-aged demographic.
*   **Flight Distance:** Flight distances vary widely, with the majority below 1743 miles. This indicates a mix of short-haul and potentially some medium-haul flights in the dataset.
*   **Service Ratings:** Ratings for inflight services, cleanliness, and comfort are distributed across the 0-5 scale, with medians generally above 3. This suggests generally positive ratings, but there's room for improvement.  Further investigation into the distribution of lower ratings is warranted.
*   **Delays:** There are some long delays, but most passengers experience minimal or no delay.  This suggests that while delays are a factor, they are not the norm.  Outliers (long delays) might need special handling.
*   **Satisfaction Rate:** The satisfaction rate is roughly 43%, indicating a class imbalance (57% of passengers are dissatisfied or neutral). This class imbalance will need to be addressed during model training to avoid biased predictions.

```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df['satisfaction'] = df['satisfaction'].apply(lambda x: 1 if x == 'satisfied' else 0)
```

```python
df = pd.get_dummies(df, columns=['Gender', 'Customer Type', 'Type of Travel', 'Class'], drop_first=True,dtype='int')
```

```python
# Correlation matrix for numerical features
corr_matrix = df.corr()

# Plotting the correlation matrix
plt.figure(figsize=(20, 10))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Matrix of Features')
plt.show()
```

**Positive Correlations with Satisfaction:** Features like Online boarding, Inflight entertainment, and Seat comfort show a moderate positive correlation with passenger satisfaction. This indicates that passengers who rate these aspects of their flight experience highly are more likely to be satisfied overall.  This is valuable information for airlines looking to improve customer satisfaction.


```python
sns.set(rc={
    "font.size":15,
    "axes.titlesize":10,
    "axes.labelsize":15},
    style="darkgrid")
fig, axs = plt.subplots(6, 3, figsize=(20,30))
fig.tight_layout(pad=4.0)

for f, ax in zip(df, axs.ravel()):
    sns.set(font_scale = 2)
    ax = sns.histplot(ax=ax, data=df, x=df[f], kde=True, color='black')
    ax.set_title(f)
```
